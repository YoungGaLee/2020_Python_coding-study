필요한 모듈을 불러온다.

```
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
```

랜덤 시드 고정.

```
np.random.seed(0)
torch.manual_seed(0)
```

데이터를 불러와서 x(data)와 y(label)로 나눈다.

```
data=pd.read_csv('data-01-test-score.csv')
scores=data.values
np.random.shuffle(scores)

data_x=list(np.delete(scores,(-1), axis=1))
data_y=list(np.array(scores[:,-1]))
print(data_x)
print(data_y)
```

normalization

```
# robust_scaler=RobustScaler()
# robust_scaler.fit(data_x)

# data_x=robust_scaler.transform(data_x)
```

전제(data)에서 train와 test 분할

```
train_x=data_x[:19]
test_x=data_x[19:]

train_y=data_y[:19]
test_y=data_y[19:]

print(test_x)
print(test_y)
```

레이어 선언, xavier 이니셜라이저로 초기화

```
linear1=nn.Linear(3,16,bias=True)
linear2=nn.Linear(16,1,bias=True)
# linear3=nn.Linear(64,1,bias=True)

relu=nn.ReLU()
```

```
torch.nn.init.xavier_uniform(linear1.weight)
torch.nn.init.xavier_uniform(linear2.weight)
# torch.nn.init.xavier_uniform(linear3.weight)
```

Train

파라미터 설정, 모델, 로스 선언

```
max_score=0
epoch=1000
lr=0.001

# model=torch.nn.Sequential(linear1)
model=torch.nn.Sequential(linear1,relu,linear2)

optimizer=torch.optim.SGD(model.parameters(), lr=lr)
criterion=nn.MSELoss()

for step in range(epoch):

  X=Variable(torch.Tensor(train_x))
  Y=Variable(torch.Tensor(train_y))

  optimizer.zero_grad()
  hypothesis=model(X)
  cost=criterion(hypothesis,Y)
  cost.backward()
  optimizer.step()

  # count=0
  # for i in range(len(test_x)):
  #   test_X=Variable(torch.Tensor(test_x[i]))
  #   hypothesis=model(test_X)
  #   h=hypothesis.detach().numpy()
  #   pred_x=np.rint(h[0])

    # if pred_x==test_y[i]:
    #   count+=1
    
  print('cost: {}   at {}epoch'.format(cost.data.numpy(),step))
```
